We'll start with the first task.

```python
def blackbody_lam(lam, T):
    """ Blackbody as a function of wavelength (um) and temperature (K).

    returns units of erg/s/cm^2/cm/Ste # convert to metres
    """
    from scipy.constants import h,k,c
    lam = 1e-6 * lam # convert to metres
    return 2*h*c**2 / (lam**5 * (np.exp(h*c / (lam*k*T)) - 1))
```

```python
lam_um = dat[:, 0]/10000.0 
BB_temp = 16258 #K
bb_spec = blackbody_lam(lam_um, BB_temp)
bb_spec_norm = bb_spec/np.mean(bb_spec)
rat = dat[:, 1]/bb_spec_norm
```

This sent me on a quest to see if pysynphot exists in astropy yet.  It doesn't, but I found a stand-alone version.  Importing it throws a lot of warnings, but I think it's OK to ignore them.

```python
import warnings
warnings.filterwarnings('ignore')

from pysynphot import observation
from pysynphot import spectrum
 
def rebin_spec(wave, specin, wavnew):
    spec = spectrum.ArraySourceSpectrum(wave=wave, flux=specin)
    f = np.ones(len(wave))
    filt = spectrum.ArraySpectralElement(wave, f, waveunits='angstrom')
    obs = observation.Observation(spec, filt, binset=wavnew, force='taper')
 
    return obs.binflux
```

Let's do a pixel-by-pixel estimate of the mean and variance of the spectrum.  We can fit with some long-tailed distributions to be robust against outliers (cosmic rays, systematic errors).  My experience on this is limited, so I'll try a few distributions and compare the results and stick with the one I like the best.

```python
from scipy import stats

n_spec = 627
specs = numpy.concatenate((ltt3218_set,ltt7987_set), axis=1)
spec_d = np.zeros((n_spec, 2))
spec_t = np.zeros((n_spec, 2))
spec_c = np.zeros((n_spec, 2))

print shape(spec_d[:,1])

for i in range(n_spec):
    spec_d[i,:] = [np.mean(specs[i, :]), np.std(specs[i, :])]
    t_shape, t_loc, t_scale = stats.t.fit(specs[i, :])
    spec_t[i,:] = [t_loc, t_scale]
```

We should perform some surgery in the vicinity of $H\alpha$.  First idea is to smooth the spectrum, then replace the errant data points with the smoothed spectrum.  Here's our chance for either Nadaraya-Watson or median smoothing.  In the process of doing this we are going to make new estimators, so let's define the Spectral Response $R_\lambda $ as  $$R_\lambda = f_{\lambda}/\hat{f}_{\lambda}$$
Then, let's just focus on the wavelength range surrounding $H\alpha$

```python
from astroML.linear_model import NadarayaWatson
x = wl_obs
y = spec_t[:,0]
dy = spec_d[:,1]
model = NadarayaWatson('gaussian', 10.0)
model.fit(x[:, None], y, dy)
y_pred = model.predict(x[:, None])

fig = plt.figure(figsize=(14, 6))

ax = fig.add_subplot(121)

ax.plot(x, y, '.k', alpha=0.5)    
ax.plot(x, y_pred, '-b')
ax.set_xlabel('$\lambda$')
ax.set_ylabel('$R_\lambda$')
ax.set_xlim((6500, 9000))

ax = fig.add_subplot(122)

ax.plot(x, y, '.k', alpha=0.5)    
ax.plot(x, y_pred, '-b')
ax.set_xlabel('$\lambda$')
ax.set_xlim((6500, 6700))
ax.set_ylim((0.65, 0.8))
ax.set_title(r'$H \alpha$ structure to be smoothed over')
ax.fill_between([6550,6580], [0.7,0.7], [0.79,0.79], alpha=0.3, color='r')

plt.show()        
```

Let's explain what it is that we're plotting above.  It is the observed telluric spectrum, $f_\lambda$, divided by the blackbody Planck spectrum $B_\lambda$ assuming an effective temperature $T$ listed in the file "tell_stds.csv", divided by the assumed average response function $R_\lambda$.  Let's define it as $\eta $ : 

$$ \eta = \frac{f_{\lambda}}{B_{\lambda}(T)} \frac{1}{R_\lambda}$$

One **key idea** is that we can construct an estimator for the intrinsic spectrum of HR5517 and HR6930 in the following way.  We'll combine the highest quality spectra of HR5517 and HR6930 from the nights for which we have near-contemporaneous similar-airmass spectra of LTT3218 and LTT7987.  Then we will smooth over any minor features of known telluric origin.  Let's drop the $\lambda$ subscripts and instead mark the observed telluric spectral flux density with $f_o$, our estimator for the "True" spectrum flux with $f_T$.

$$\hat{f}_{True} = \hat{\eta}B(T)$$

This is only *sort of* cheating, and *sort of* bad.  There are several issues making this strategy sub-optimal.  Chiefly, we are adding twice the noise than we have to (noise from eta-hat and noise from our telluric observation).  There are also the concerns of "do we have the black body effective temperature right?" and "how good is our estimate $\hat{\eta}$, and do telluric features creep into it?"  My intuition says that all these issues matter at such a small level that we don't care, and can just move on.  The Right Thing to Do would be to match exactly what $\tt{xTellCorr}$ from $\tt{SpeXtool}$ does-- namely, go out and find a model spectrum of the same spectral type, then convolve the spectral lines of the model to the match the doppler broadened observed lines.  There's also a B-V correction to do.  All of that would require hacking SpeXtool, which although noble, would take longer to do than what I'm doing now since it would require debugging someone else's code.

It's important to emphasize that this strategy only works in comparison to response functions derived with near-contemporaneous similar-airmass observations.  You can tell by looking at the figure above that observations from other nights and years look very different.  That's exactly why we wish to construct the estimator $\hat{f}_{True}$ so that we can apply it to those other ones.  Whew.  OK here we go.
